```python
# ─────────────────────────────────────────────────────────────────────────────
# SECTION 1 │ BUILT-IN FUNCTIONS
# ─────────────────────────────────────────────────────────────────────────────

print("=" * 60)
print("SECTION 1 — BUILT-IN FUNCTIONS")
print("=" * 60)

# ── Example 1a: map() ─────────────────────────────────────────────────────────
# Apply a function to EVERY item in an iterable.
# Returns a lazy map object — wrap with list() to see it.

print("\n▸ Example 1a — map()")

prices     = [12.5, 30.0, 8.75, 45.0]
discounted = list(map(lambda p: round(p * 0.80, 2), prices))

print(f"  Original      : {prices}")
print(f"  After 20% off : {discounted}")
# → [10.0, 24.0, 7.0, 36.0]

# ── Example 1b: filter() ──────────────────────────────────────────────────────
# Keep only items where the predicate returns True.

print("\n▸ Example 1b — filter()")

scores  = [55, 88, 42, 71, 60, 39, 95]
passing = list(filter(lambda s: s >= 60, scores))
failing  = list(filter(lambda s: s <  60, scores))

print(f"  All scores : {scores}")
print(f"  Passing    : {passing}")
print(f"  Failing    : {failing}")

# ── Example 1c: zip() ─────────────────────────────────────────────────────────
# Pair elements from two or more iterables; stops at the shortest.

print("\n▸ Example 1c — zip()")

students = ["Alice", "Bob", "Cara"]
grades   = ["A",    "B+",  "A-"]

for name, grade in zip(students, grades):
    print(f"  {name}: {grade}")

lookup = dict(zip(students, [95, 87, 91]))
print(f"  Score lookup dict: {lookup}")

# ── Example 1d: enumerate() ───────────────────────────────────────────────────
# Add an automatic counter; yields (index, value) pairs.

print("\n▸ Example 1d — enumerate()")

menu = ["Coffee", "Tea", "Juice", "Water"]
for number, item in enumerate(menu, start=1):
    print(f"  {number}. {item}")
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# SECTION 3 │ ERROR HANDLING
# ─────────────────────────────────────────────────────────────────────────────

print("\n" + "=" * 60)
print("SECTION 3 — ERROR HANDLING")
print("=" * 60)

# ── Example 3a — try / except with multiple exception types ───────────────────
print("\n▸ Example 3a — safe integer conversion")

def safe_int(value):
    try:
        return int(value)
    except ValueError:
        print(f"  ValueError: '{value}' cannot be converted to int.")
        return None
    except TypeError:
        print(f"  TypeError: cannot convert {type(value).__name__}.")
        return None

for v in ["42", "hello", None, "007"]:
    result = safe_int(v)
    print(f"  safe_int({str(v)!r:>8}) → {result}")

# ── Example 3b — try / except / else / finally ────────────────────────────────
print("\n▸ Example 3b — try / except / else / finally")

def read_first_line(path):
    try:
        f = open(path, "r")
        line = f.readline()
    except FileNotFoundError:
        print(f"  File '{path}' not found.")
        return None
    else:
        f.close()
        print(f"  Read OK → {line.strip()!r}")
        return line.strip()
    finally:
        print("  (finally: always runs)")

read_first_line("nonexistent.txt")
```

## **PROBLEM 1 │ GRADE REPORTER**

```python
# ─────────────────────────────────────────────────────────────────────────────
# PROBLEM 1 │ GRADE REPORTER
# Concepts  : map · filter · enumerate
# ─────────────────────────────────────────────────────────────────────────────
print("\n── Problem 1: Grade Reporter ─────────────────────────────────────────")
"""
Given (name, score) pairs, produce a numbered report with letter grades
and separate passing from failing students.
"""

records = [
    ("Alice", 92), ("Bob", 55), ("Cara", 78),
    ("Dan",   61), ("Eve", 45), ("Frank", 88),
]

def to_grade(score):
    return ("A" if score >= 90 else "B" if score >= 80 else
            "C" if score >= 70 else "D" if score >= 60 else "F")


# map: attach grade; filter: split pass/fail; enumerate: numbered output
graded  = list(map(lambda r: (*r, to_grade(r[1])), records))
passing = list(filter(lambda r: r[1] >= 60, graded))
failing  = list(filter(lambda r: r[1] <  60, graded))

print(f"\n  {'#':<4}{'Name':<8}{'Score':>6}{'Grade':>7}")
print(f"  {'-'*26}")
for idx, (name, score, grade) in enumerate(graded, 1):
    print(f"  {idx:<4}{name:<8}{score:>6}{grade:>7}  {'✓' if score >= 60 else '✗'}")

avg = sum(r[1] for r in records) / len(records)
print(f"\n  Average: {avg:.1f}  |  Passing: {len(passing)}  |  Failing: {len(failing)}")
```

## **PROBLEM 2 │ CSV LOG FILE ANALYSER**

```python
# ─────────────────────────────────────────────────────────────────────────────
# PROBLEM 2 │ CSV LOG FILE ANALYSER
# Concepts  : File I/O · try-except · filter · enumerate
# ─────────────────────────────────────────────────────────────────────────────
print("\n── Problem 2: CSV Log Analyser ───────────────────────────────────────")
"""
Parse an access log CSV; skip headers and malformed lines;
report all HTTP error entries (status >= 400).
"""

LOG = "access.log"
raw_log = [
    "timestamp,endpoint,status",
    "2024-01-01 08:00,/api/login,200",
    "2024-01-01 08:01,/api/data,500",
    "CORRUPTED LINE",
    "2024-01-01 08:02,/api/login,401",
    "2024-01-01 08:03,/api/home,200",
    "2024-01-01 08:04,/api/delete,403",
]

with open(LOG, "w") as f:
    f.write("\n".join(raw_log) + "\n")

entries, bad = [], []

with open(LOG, "r") as f:
    for lineno, raw in enumerate(f, 1):
        line = raw.strip()
        if lineno == 1 or not line:
            continue
        try:
            ts, endpoint, status = line.split(",")
            entries.append({"ts": ts, "ep": endpoint, "status": int(status)})
        except ValueError:
            bad.append((lineno, line))

errors = list(filter(lambda e: e["status"] >= 400, entries))

print(f"\n  Valid: {len(entries)}  |  Bad lines: {len(bad)}  |  Errors: {len(errors)}")
print("\n  HTTP Errors:")
for e in errors:
    print(f"    [{e['status']}] {e['ts']}  {e['ep']}")
if bad:
    print(f"\n  Malformed lines:")
    for ln, txt in bad:
        print(f"    Line {ln}: {txt!r}")

os.remove(LOG)
```

## **PROBLEM 3 │ FILE-BACKED CONTACT BOOK**

```python
# ─────────────────────────────────────────────────────────────────────────────
# PROBLEM 3 │ FILE-BACKED CONTACT BOOK
# Concepts  : File I/O · zip · filter · try-except
# ─────────────────────────────────────────────────────────────────────────────
print("\n── Problem 3: Contact Book ───────────────────────────────────────────")
"""
Persist contacts to a pipe-delimited file. Load with zip(),
search by keyword using filter().
"""

CONTACTS = "contacts.txt"
FIELDS   = ["name", "phone", "email"]

raw_contacts = [
    ("Alice",  "555-0101", "alice@example.com"),
    ("Bob",    "555-0102", "bob@work.com"),
    ("Cara",   "555-0103", "cara@example.com"),
    ("Dan",    "555-0104", "dan@mail.net"),
]

with open(CONTACTS, "w") as f:
    for row in raw_contacts:
        f.write("|".join(row) + "\n")

def load_contacts(path):
    contacts = []
    try:
        with open(path) as f:
            for line in f:
                parts = line.strip().split("|")
                if len(parts) == 3:
                    contacts.append(dict(zip(FIELDS, parts)))   # zip here
    except FileNotFoundError:
        print("  Contact file not found.")
    return contacts

def search_contacts(contacts, query):
    q = query.lower()
    return list(filter(
        lambda c: q in c["name"].lower() or q in c["email"].lower(),
        contacts
    ))

loaded  = load_contacts(CONTACTS)
results = search_contacts(loaded, "example")

print(f"\n  Loaded {len(loaded)} contacts. Search 'example' → {len(results)} match(es):")
for c in results:
    print(f"    {c['name']:<8} {c['phone']:<12} {c['email']}")

os.remove(CONTACTS)
```

## **PROBLEM 4 │ SAFE CALCULATOR**

```python
# ─────────────────────────────────────────────────────────────────────────────
# PROBLEM 4 │ SAFE CALCULATOR
# Concepts  : try-except · map · operator module · input validation
# ─────────────────────────────────────────────────────────────────────────────
print("\n── Problem 4: Safe Calculator ────────────────────────────────────────")
"""
Parse "num op num" expressions safely. Never crash — return a clear
error message for every invalid input.
"""

import operator as op

OPS = {"+": op.add, "-": op.sub, "*": op.mul, "/": op.truediv,
       "//": op.floordiv, "%": op.mod, "**": op.pow}

def calculate(expr):
    try:
        parts = expr.strip().split()
        if len(parts) != 3:
            return None, "Format must be: <num> <op> <num>"

        left_s, op_s, right_s = parts

        try:
            left, right = float(left_s), float(right_s)
        except ValueError:
            return None, f"'{left_s}' or '{right_s}' is not a number"

        if op_s not in OPS:
            return None, f"Unknown operator '{op_s}'"

        if op_s in ("/", "//", "%") and right == 0:
            return None, "Division by zero"

        result = OPS[op_s](left, right)
        return (int(result) if result == int(result) else round(result, 6)), None

    except Exception as e:
        return None, str(e)

tests = ["10 + 5", "100 / 4", "7 / 0", "2 ** 8", "abc + 3", "9 // 2"]

print(f"\n  {'Expression':<18}  Result")
print(f"  {'-'*35}")
for expr in tests:
    val, err = calculate(expr)
    print(f"  {expr:<18}  {'= ' + str(val) if val is not None else '✗ ' + err}")
```

## **PROBLEM 5 │ WORD FREQUENCY COUNTER**

```python
# ─────────────────────────────────────────────────────────────────────────────
# PROBLEM 5 │ WORD FREQUENCY COUNTER
# Concepts  : File I/O · list comprehension · enumerate · sort · try-except
# ─────────────────────────────────────────────────────────────────────────────
print("\n── Problem 5: Word Frequency Counter ────────────────────────────────")
"""
Read a text file, tokenise, drop stop-words, count frequencies,
rank top-10 and display with a visual bar chart.
"""

TEXT = "article.txt"
sample_text = (
    "Python is a versatile programming language widely used in web development, "
    "data science, and automation. Python is easy to learn and powerful enough "
    "for professional software development. Many companies use Python for machine "
    "learning, data analysis, and building APIs. Python is growing rapidly."
)

with open(TEXT, "w") as f:
    f.write(sample_text)

STOPWORDS = {"a","an","the","is","in","and","for","to","of","are",
             "it","be","as","at","by","or","on","was","with","that","this"}

def word_frequencies(path, top_n=10):
    try:
        with open(path) as f:
            text = f.read()
    except FileNotFoundError:
        return []

    tokens = [w.strip(string.punctuation).lower() for w in text.split()]
    words  = [w for w in tokens if w and w not in STOPWORDS and len(w) > 2]

    freq = {}
    for w in words:
        freq[w] = freq.get(w, 0) + 1

    return sorted(freq.items(), key=lambda kv: kv[1], reverse=True)[:top_n]

ranking = word_frequencies(TEXT)
max_cnt = ranking[0][1] if ranking else 1

print(f"\n  {'#':<4}{'Word':<18}{'Count':>5}  Bar")
print(f"  {'-'*44}")
for rank, (word, count) in enumerate(ranking, 1):
    bar = "█" * int((count / max_cnt) * 18)
    print(f"  {rank:<4}{word:<18}{count:>5}  {bar}")

os.remove(TEXT)

print("\n" + "=" * 60)
print("  Code Lab complete — great work!")
print("=" * 60)
```
