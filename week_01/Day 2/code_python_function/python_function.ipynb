# **What is a Function?**

A reusable block of code that performs a specific task.
Functions help you:

* Break down complex problems
* Avoid code repetition (DRY principle)
* Make code easier to test and debug
* Improve code readability

### **Example 1.2: Functions with Return Values**
* Demonstrate the difference between printing and returning

**Points:**

* Explain that return gives data back to the caller
* Show that returned values can be stored, modified, or reused
* Emphasize that print is for output, return is for passing data
* Demonstrate that functions without return implicitly return None

```python
# Function that prints (side effect)
def greet_print(name):
    """Print a greeting (side effect)."""
    print(f"Hello, {name}!")


# Function that returns (pure function)
def greet_return(name):
    """Return a greeting message.

    Args:
        name: The name to greet

    Returns:
        A greeting string
    """
    return f"Hello, {name}!"

# Demonstrate the difference
print("=== Using print function ===")
greet_print("Alice")
# Can't do anything with the result - it's already printed

print("\n=== Using return function ===")
message = greet_return("Bob")
print(message)  # We control when to print
print(message.upper())  # We can manipulate it
# Can use it multiple times
print(f"Saved message: {message}")
```

### **Example 1.3: Building a Calculator - Multiple Parameters**
* Show practical use of parameters and mathematical operations


**Key Points:**
* Show how parameters make functions flexible
* Demonstrate error handling (division by zero)
* Explain that returning None can signal an error
* Show consistent function naming (all lowercase verbs)

```python
def add(a, b):
    """Add two numbers.

    Args:
        a: First number
        b: Second number

    Returns:
        Sum of a and b
    """
    return a + b

def subtract(a, b):
    """Subtract b from a."""
    return a - b

def multiply(a, b):
    """Multiply two numbers."""
    return a * b

def divide(a, b):
    """Divide a by b.

    Args:
        a: Numerator
        b: Denominator

    Returns:
        Result of division, or None if b is zero
    """
    if b == 0:
        print("Error: Cannot divide by zero!")
        return None
    return a / b

# Using the calculator functions
print("Addition: 10 + 5 =", add(10, 5))
print("Subtraction: 10 - 5 =", subtract(10, 5))
print("Multiplication: 10 * 5 =", multiply(10, 5))
print("Division: 10 / 5 =", divide(10, 5))
print("Division by zero:", divide(10, 0))
```

## **Part 2: Parameters - Making Functions Flexible**

### **Example 2.1: Default Parameters**
* Show how defaults make functions more convenient

**Key Points:**

* Required parameters come before default parameters
* Default values are set once when function is defined
* Keyword arguments make code more readable
* Show practical use case (user account creation)

```python
def create_user(username, role="user", active=True):
    """Create a user account with default settings.

    Args:
        username: User's username (required)
        role: User role (default: "user")
        active: Account active status (default: True)

    Returns:
        Dictionary with user information
    """
    user = {
        "username": username,
        "role": role,
        "active": active
    }
    return user

# Using defaults
user1 = create_user("alice")
print("User 1:", user1)

# Overriding one default
user2 = create_user("bob", role="admin")
print("User 2:", user2)

# Overriding all defaults
user3 = create_user("charlie", role="moderator", active=False)
print("User 3:", user3)

# Using keyword arguments (order doesn't matter)
user4 = create_user(active=False, username="diana", role="guest")
print("User 4:", user4)
```

### Example 2.2: Building a Flexible Formatter
* Show real-world use of multiple default parameters

**Key Points:**

* *args collects arguments into a tuple
* Useful for functions that work with variable-length inputs
* Can combine with regular and default parameters
* Show practical uses (averaging, totals, etc.)

```python
def calculate_average(*numbers):
    """Calculate average of any number of values.

    Args:
        *numbers: Variable number of numeric arguments

    Returns:
        Average of all numbers, or 0 if none provided
    """
    if len(numbers) == 0:
        return 0

    total = sum(numbers)
    average = total / len(numbers)
    return average

# Can call with any number of arguments
print("Average of 5, 10 is:", calculate_average(5, 10))
print("Average of 1, 2, 3, 4, 5 is:", calculate_average(1, 2, 3, 4, 5))
print("Average of 100 is:", calculate_average(100))
print("Average of nothing is:", calculate_average())
```

```python
# Practical example: shopping cart total
def calculate_total(*prices, tax_rate=0.08):
    """Calculate total with tax for multiple items.

    Args:
        *prices: Variable number of item prices
        tax_rate: Tax rate as decimal (default: 0.08)

    Returns:
        Total price with tax
    """
    subtotal = sum(prices)
    tax = subtotal * tax_rate
    return subtotal + tax

print("\nShopping cart:")
total = calculate_total(19.99, 29.99, 9.99)
print(f"Total: ${total:.2f}")
```

## **Part 3: Return Values - Getting Data Back**

### **Example 3.1: Multiple Return Values**
* Show how to return multiple pieces of data

**Key Points:**

* Multiple values are returned as a tuple
* Can unpack into separate variables
* Or keep as a single tuple
* Show validation (empty list case)

```python
def get_user_stats(scores):
    """Calculate statistics for user scores.

    Args:
        scores: List of numeric scores

    Returns:
        Tuple of (min, max, average)
    """
    if not scores:
        return 0, 0, 0

    min_score = min(scores)
    max_score = max(scores)
    avg_score = sum(scores) / len(scores)

    return min_score, max_score, avg_score

# Using multiple return values
scores = [85, 92, 78, 95, 88]
minimum, maximum, average = get_user_stats(scores)

print(f"Scores: {scores}")
print(f"Minimum: {minimum}")
print(f"Maximum: {maximum}")
print(f"Average: {average:.2f}")

# Can also return as tuple
stats = get_user_stats(scores)
print(f"\nAs tuple: {stats}")
print(f"Type: {type(stats)}")
```

### **Example 3.2: Returning Different Types**
* Demonstrate flexibility in return types

**Key Points:**

* Functions can return different types based on situation
* Tuple return is useful for (success, result/error) pattern
* Demonstrate boolean logic with validation
* Show practical error handling

```python
def validate_password(password):
    """Validate a password against security rules.

    Args:
        password: The password to validate

    Returns:
        Tuple of (is_valid, error_message)
        is_valid is True if password is valid, False otherwise
        error_message is None if valid, otherwise describes the error
    """
    if len(password) < 8:
        return False, "Password must be at least 8 characters"

    if not any(char.isupper() for char in password):
        return False, "Password must contain at least one uppercase letter"

    if not any(char.isdigit() for char in password):
        return False, "Password must contain at least one digit"

    return True, None

# Testing different passwords
passwords = ["short", "alllowercase123", "NoDigits", "ValidPass123"]

for pwd in passwords:
    is_valid, error = validate_password(pwd)

    if is_valid:
        print(f"✓ '{pwd}' is valid")
    else:
        print(f"✗ '{pwd}' is invalid: {error}")
```

### **Example 3.3: Early Returns for Clarity**
* Show how early returns simplify logic

**Key Points:**

* Early returns avoid deep nesting
* Check edge cases first
* Each return is clear and specific
* More readable than complex if-elif-else chains

```python
def calculate_discount(price, customer_type, quantity):
    """Calculate discount based on customer type and quantity.

    Args:
        price: Base price per item
        customer_type: "regular", "premium", or "vip"
        quantity: Number of items

    Returns:
        Final price after discounts
    """
    # Early return for invalid input
    if price <= 0 or quantity <= 0:
        return 0

    total = price * quantity

    # VIP: 25% discount
    if customer_type == "vip":
        return total * 0.75

    # Premium: 15% discount
    if customer_type == "premium":
        return total * 0.85

    # Regular: bulk discount only (10% for 10+ items)
    if quantity >= 10:
        return total * 0.90

    # No discount
    return total

# Test different scenarios
print("Regular, 5 items:", calculate_discount(10, "regular", 5))
print("Regular, 15 items:", calculate_discount(10, "regular", 15))
print("Premium, 5 items:", calculate_discount(10, "premium", 5))
print("VIP, 5 items:", calculate_discount(10, "vip", 5))
print("Invalid:", calculate_discount(-10, "regular", 5))
```

## **Example 4.1: Complete Docstring Template**

* Show professional documentation standards

# Key Points:

* Complete docstrings include description, args, returns, exceptions
* Type hints in docstrings help users
* Examples in docstrings are very helpful
* help() function displays the docstring

```python
def process_transaction(account_id, amount, transaction_type="debit"):
    """Process a financial transaction for an account.

    This function validates the transaction, updates the account balance,
    and returns the transaction result.

    Args:
        account_id (str): Unique identifier for the account
        amount (float): Transaction amount in dollars
        transaction_type (str): Either "debit" or "credit" (default: "debit")

    Returns:
        dict: Transaction result containing:
            - success (bool): Whether transaction succeeded
            - new_balance (float): Updated account balance
            - message (str): Status message

    Raises:
        ValueError: If amount is negative or transaction_type is invalid

    Examples:
        >>> process_transaction("ACC123", 50.00, "credit")
        {'success': True, 'new_balance': 1050.00, 'message': 'Credit processed'}

        >>> process_transaction("ACC123", 2000.00, "debit")
        {'success': False, 'new_balance': 1050.00, 'message': 'Insufficient funds'}
    """
    # Validation
    if amount < 0:
        raise ValueError("Amount cannot be negative")

    if transaction_type not in ["debit", "credit"]:
        raise ValueError("Transaction type must be 'debit' or 'credit'")

    # Simulated account balance
    current_balance = 1000.00

    if transaction_type == "credit":
        new_balance = current_balance + amount
        return {
            "success": True,
            "new_balance": new_balance,
            "message": "Credit processed"
        }
    else:  # debit
        if current_balance >= amount:
            new_balance = current_balance - amount
            return {
                "success": True,
                "new_balance": new_balance,
                "message": "Debit processed"
            }
        else:
            return {
                "success": False,
                "new_balance": current_balance,
                "message": "Insufficient funds"
            }

# Using the function
result = process_transaction("ACC123", 50.00, "credit")
print(result)

# View the docstring
print("\n" + "="*50)
print("Function documentation:")
print("="*50)
help(process_transaction)
```

### **Example 4.2: Accessing Docstrings Programmatically**
* Show that docstrings are accessible at runtime

**Key Points:**

* `__doc__` attribute contains the docstring
* Docstrings are part of the function object
* IDEs use docstrings for autocomplete
* Professional practice for library code

```python
def celsius_to_fahrenheit(celsius):
    """Convert Celsius to Fahrenheit.

    Args:
        celsius (float): Temperature in Celsius

    Returns:
        float: Temperature in Fahrenheit
    """
    return (celsius * 9/5) + 32

def fahrenheit_to_celsius(fahrenheit):
    """Convert Fahrenheit to Celsius.

    Args:
        fahrenheit (float): Temperature in Fahrenheit

    Returns:
        float: Temperature in Celsius
    """
    return (fahrenheit - 32) * 5/9

# Access docstrings programmatically
print("Function: celsius_to_fahrenheit")
print("Documentation:", celsius_to_fahrenheit.__doc__)
print()

print("Function: fahrenheit_to_celsius")
print("Documentation:", fahrenheit_to_celsius.__doc__)
print()

# Use the functions
temp_c = 25
temp_f = celsius_to_fahrenheit(temp_c)
print(f"{temp_c}°C = {temp_f}°F")

temp_f = 77
temp_c = fahrenheit_to_celsius(temp_f)
print(f"{temp_f}°F = {temp_c:.1f}°C")
```

### **Example 5.1: Understanding Local vs Global Scope**
* Clearly demonstrate scope boundaries

**Key Points:**

* Global variables are accessible everywhere
* Need global keyword to modify global variables
* Local variables exist only within their function
* Accessing local variables outside their function causes NameError

```python
# Global variable
player_score = 0

def add_points(points):
    """Add points to global score using global keyword."""
    global player_score
    player_score += points
    print(f"Added {points} points. New score: {player_score}")

def calculate_bonus(multiplier):
    """Calculate bonus - demonstrates local variable."""
    # This is a LOCAL variable, different from global player_score
    bonus_score = player_score * multiplier
    print(f"Bonus calculation: {player_score} × {multiplier} = {bonus_score}")
    return bonus_score

# Using the functions
print("Initial score:", player_score)
add_points(10)
add_points(5)
bonus = calculate_bonus(2)
print(f"Bonus: {bonus}")
print(f"Final score: {player_score}")
```

### **Example 5.2: Best Practice - Avoiding Globals**
* Show better alternative to global variables

**Key Points:**

* Global variables make code harder to test and debug
* Pure functions (no side effects) are easier to reason about
* Pass data through parameters, return new values
* More predictable and reusable

```python
# ❌ BAD: Using global variables
score = 0

def bad_add_points(points):
    """Bad practice: modifying global variable."""
    global score
    score += points

# ✓ GOOD: Using parameters and returns
def good_add_points(current_score, points):
    """Good practice: pure function with parameters and return.

    Args:
        current_score: Current score value
        points: Points to add

    Returns:
        New score after adding points
    """
    return current_score + points

# Demonstrating the difference
print("=== Bad approach (global) ===")
bad_add_points(10)
bad_add_points(5)
print(f"Score: {score}")

print("\n=== Good approach (parameters) ===")
player_score = 0
player_score = good_add_points(player_score, 10)
player_score = good_add_points(player_score, 5)
print(f"Score: {player_score}")
```

### **Example 5.3: Scope in Nested Functions**
* Show enclosing scope (for advanced students)

**Key Points:**

* Inner functions can access outer function variables
* This creates a "closure" that remembers the outer variable
* Advanced topic but demonstrates scope hierarchy
* Practical use: factory functions

```python
def create_multiplier(factor):
    """Create a function that multiplies by a specific factor.

    This demonstrates closures and enclosing scope.

    Args:
        factor: The multiplication factor

    Returns:
        A function that multiplies its argument by factor
    """
    def multiply(number):
        """Multiply number by the enclosing factor."""
        return number * factor

    return multiply

# Create specialized multiplier functions
double = create_multiplier(2)
triple = create_multiplier(3)
times_ten = create_multiplier(10)

# Use them
print("Double 5:", double(5))
print("Triple 5:", triple(5))
print("Times ten 5:", times_ten(5))

# Demonstrate that each function "remembers" its factor
print("\nEach function has its own factor:")
print(f"double(10) = {double(10)}")
print(f"triple(10) = {triple(10)}")
print(f"times_ten(10) = {times_ten(10)}")
```

### **Example 6.1: Student Grade Manager (Complete System)**
* Demonstrate integration of all concepts

**Key Points:**

* Each function has a single, clear purpose
* Functions work together to build complex functionality
* Well-documented with docstrings
* Easy to test individual functions
* Professional code structure

```python
def calculate_letter_grade(score):
    """Convert numeric score to letter grade.

    Args:
        score (float): Numeric score 0-100

    Returns:
        str: Letter grade (A, B, C, D, or F)
    """
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

def calculate_average(scores):
    """Calculate average score.

    Args:
        scores (list): List of numeric scores

    Returns:
        float: Average score, or 0 if list is empty
    """
    if not scores:
        return 0
    return sum(scores) / len(scores)

def get_student_report(name, scores):
    """Generate complete student report.

    Args:
        name (str): Student name
        scores (list): List of test scores

    Returns:
        dict: Report containing name, scores, average, and grade
    """
    average = calculate_average(scores)
    grade = calculate_letter_grade(average)

    return {
        "name": name,
        "scores": scores,
        "average": round(average, 2),
        "letter_grade": grade,
        "num_tests": len(scores)
    }

def display_report(report):
    """Display formatted student report.

    Args:
        report (dict): Student report from get_student_report
    """
    print(f"\n{'='*40}")
    print(f"Student Report: {report['name']}")
    print(f"{'='*40}")
    print(f"Number of tests: {report['num_tests']}")
    print(f"Scores: {', '.join(map(str, report['scores']))}")
    print(f"Average: {report['average']}")
    print(f"Letter Grade: {report['letter_grade']}")
    print(f"{'='*40}")

# Using the complete system
alice_scores = [85, 92, 78, 95, 88]
bob_scores = [72, 75, 70, 78]

alice_report = get_student_report("Alice Johnson", alice_scores)
bob_report = get_student_report("Bob Smith", bob_scores)

display_report(alice_report)
display_report(bob_report)
```

### **Example 6.2: Input Validation Library**
* Build a reusable validation toolkit

**Key Points:**

* Building a reusable library of functions
* Consistent return format across functions
* Composing simple functions into complex ones
* Error messages help users fix problems
* Professional validation patterns

```python
def validate_email(email):
    """Validate email format.

    Args:
        email (str): Email address to validate

    Returns:
        tuple: (is_valid, error_message)
    """
    if "@" not in email:
        return False, "Email must contain @"

    if email.count("@") > 1:
        return False, "Email can only have one @"

    username, domain = email.split("@")

    if not username:
        return False, "Email must have username before @"

    if not domain or "." not in domain:
        return False, "Email must have valid domain"

    return True, None

def validate_phone(phone):
    """Validate phone number format (US format).

    Args:
        phone (str): Phone number to validate

    Returns:
        tuple: (is_valid, error_message)
    """
    # Remove common formatting
    cleaned = phone.replace("-", "").replace("(", "").replace(")", "").replace(" ", "")

    if not cleaned.isdigit():
        return False, "Phone must contain only digits and formatting"

    if len(cleaned) != 11:
        return False, "Phone must be 11 digits"

    return True, None

def validate_age(age):
    """Validate age is reasonable.

    Args:
        age: Age value to validate (can be int or str)

    Returns:
        tuple: (is_valid, error_message)
    """
    try:
        age_int = int(age)
    except (ValueError, TypeError):
        return False, "Age must be a number"

    if age_int < 0:
        return False, "Age cannot be negative"

    if age_int > 120:
        return False, "Age seems unrealistic"

    return True, None

def validate_user_input(email, phone, age):
    """Validate all user inputs and return combined results.

    Args:
        email (str): Email address
        phone (str): Phone number
        age: Age value

    Returns:
        dict: Validation results for each field
    """
    results = {
        "email": validate_email(email),
        "phone": validate_phone(phone),
        "age": validate_age(age)
    }

    all_valid = all(valid for valid, _ in results.values())

    return {
        "all_valid": all_valid,
        "results": results
    }

# Testing the validation library
test_data = [
    ("alice@example.com", "0855-123-4567", 25),
    ("invalid.email", "0715-1234", 200),
    ("bob@test.co", "5551234567", "thirty"),
]

for email, phone, age in test_data:
    print(f"\nValidating: {email}, {phone}, {age}")
    validation = validate_user_input(email, phone, age)

    if validation["all_valid"]:
        print("✓ All fields valid!")
    else:
        print("✗ Validation errors:")
        for field, (valid, error) in validation["results"].items():
            if not valid:
                print(f"  - {field}: {error}")
```
