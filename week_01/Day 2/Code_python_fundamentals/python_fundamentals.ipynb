##### String Slicing and Reassembling

Rule: string[start : stop : step]

```python
text = 'Linode! System'

print(text[:5]) # Start at beginning, stop before index 5. However, index 5 is excluded.
print(text[::-1]) # Reverse the string : -1 step → move backwards
print(text[3::2]) # Start at index 3, take every 2nd character
```

```python
text_02 = 'Homogeneous'

# Basic Slicing
basic_slice = text_02[4:8]
print(basic_slice)

# Step Slicing
step_slice = text_02[::2]
print(step_slice)

# Negative Slicing
neg_slice = text_02[:-3]
print(neg_slice )

# Reverse Slicing
rev_slice = text_02[::-1]
print(rev_slice)
```

#### Example 2.4: String Formatting Comparison

**Key Points:**

Three Modern Approaches

```python
name = "Alice"
age = 25
score = 87.654321

print("=== String Formatting Methods ===") # Recommended

# Method 1: f-strings (recommended)
message1 = f"My name is {name}, I am {age} years old, and my score is {score:.2f}"
print(f"f-string: {message1}")
```

```python
# Basic while loop
print("=== Countdown ===")
count = 5
while count > 0:
    print(count)
    count -= 1
print("Blast off!")
```

```python
# While loop with user input
print("\n=== Number Guessing Game ===")
secret_number = 7
attempts = 0
max_attempts = 5

while attempts < max_attempts:
    guess = int(input(f"Guess the number (1-10), attempt {attempts + 1}/{max_attempts}: "))
    attempts += 1
    
    if guess == secret_number:
        print(f"Correct! You won in {attempts} attempts!")
        break
    elif guess < secret_number:
        print("Too low!")
    else:
        print("Too high!")
else:
    print(f"Game over! The number was {secret_number}")
```

```python
# Using continue to skip
print("\n=== Even numbers only ===")
num = 0
while num < 10:
    num += 1
    if num % 2 != 0:  # If odd
        continue  # Skip to next iteration
    print(num, end=" ")
print()
```

```python
# Practical example: Input validation
print("\n=== Age Input Validation ===")
while True:
    age_input = input("Enter your age (1-120): ")
    
    if not age_input.isdigit():
        print("Please enter a valid number.")
        continue
    
    age = int(age_input)
    
    if age < 1 or age > 120:
        print("Age must be between 1 and 120.")
        continue
    
    print(f"Thank you! Your age is {age}")
    break
```

#### **Example 3.5: Loop Comparison and Best Practices**
* Show when to use for vs while
* Demonstrate menu-driven programs
* Show how to build lists with user input
* Emphasize choosing the right loop type

```python
# Example: Processing a list
colors = ["red", "green", "blue", "yellow"]

# Using for loop (preferred for known sequences)
print("=== Using for loop ===")
for color in colors:
    print(f"Color: {color}")

# Using while loop (less pythonic for this case)
print("\n=== Using while loop ===")
index = 0
while index < len(colors):
    print(f"Color: {colors[index]}")
    index += 1
```

```python
# When while is appropriate: unknown iterations
print("\n=== Reading until empty input ===")
entries = []
print("Enter items (press Enter to finish):")
while True:
    item = input("> ")
    if item == "":
        break
    entries.append(item)

print(f"\nYou entered {len(entries)} items:")
for item in entries:
    print(f"- {item}")
```

#### **Integrated Example: Student Grade Manager**

**Key Points:**
* This example combines all concepts learned
* Shows real-world application structure
* Demonstrates data organization with dictionaries
* Shows input validation, loops, conditionals, and string formatting
* Great for end-of-lesson demonstration

```python
## Student Grade Manager - Integrated Example

print("=" * 50)
print("STUDENT GRADE MANAGER".center(50))
print("=" * 50)

# Initialize data
students = []

# Menu loop
while True:
    print("\n--- MENU ---")
    print("1. Add student")
    print("2. View all students")
    print("3. Calculate class average")
    print("4. Exit")
    
    choice = input("\nEnter choice (1-4): ")
    
    if choice == "1":
        # Add student
        name = input("Enter student name: ").strip().title()
        
        # Validate score
        while True:
            score_input = input("Enter score (0-100): ")
            if score_input.isdigit():
                score = int(score_input)
                if 0 <= score <= 100:
                    break
            print("Invalid score. Please enter a number between 0 and 100.")
        
        # Determine grade
        if score >= 90:
            grade = "A"
        elif score >= 80:
            grade = "B"
        elif score >= 70:
            grade = "C"
        elif score >= 60:
            grade = "D"
        else:
            grade = "F"
        
        # Store student data
        student = {
            "name": name,
            "score": score,
            "grade": grade
        }
        students.append(student)
        
        print(f"\n✓ Added {name} - Score: {score}, Grade: {grade}")
    
    elif choice == "2":
        # View all students
        if len(students) == 0:
            print("\nNo students in the system.")
        else:
            print(f"\n{'Name':<20} {'Score':>10} {'Grade':>10}")
            print("-" * 40)
            for student in students:
                print(f"{student['name']:<20} {student['score']:>10} {student['grade']:>10}")
    
    elif choice == "3":
        # Calculate average
        if len(students) == 0:
            print("\nNo students to calculate average.")
        else:
            total = 0
            for student in students:
                total += student['score']
            
            average = total / len(students)
            print(f"\nClass average: {average:.2f}")
            print(f"Total students: {len(students)}")
            
            # Count grades
            grade_counts = {"A": 0, "B": 0, "C": 0, "D": 0, "F": 0}
            for student in students:
                grade_counts[student['grade']] += 1
            
            print("\nGrade Distribution:")
            for grade, count in grade_counts.items():
                if count > 0:
                    percentage = (count / len(students)) * 100
                    print(f"  {grade}: {count} students ({percentage:.1f}%)")
    
    elif choice == "4":
        print("\nThank you for using Student Grade Manager!")
        break
    
    else:
        print("\nInvalid choice. Please enter 1-4.")

print("\nProgram terminated.")
```
