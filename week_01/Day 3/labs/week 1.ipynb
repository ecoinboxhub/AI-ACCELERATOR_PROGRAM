```python
# Example list of colors
colors = ['red', 'green', 'blue', 'yellow']

# Accessing elements with positive indexing
print("Positive Indexing:")
print(colors[0])  # first element
print(colors[1])
print(colors[2])
```

```python
# Accessing elements with negative indexing
print("\nNegative Indexing:")
print(colors[-1])  # last element
print(colors[-2])
print(colors[-3])
```

colors = ['red', 'green', 'blue', 'yellow']

```python
# Slicing lists: extracting subsets
print("\nList Slicing:")
print(colors[1:3])  # from index 1 to 2
print(colors[:2])   # first two elements
print(colors[::2])  # every second element
print(colors[::-1]) # reverse the list
```

```python
numbers = [1, 2, 3]
numbers.append(4)  # add at the end
print("After append:", numbers)
numbers.insert(1, 5)  # insert at index 1
print("After insert:", numbers)
numbers.remove(2)     # remove first occurrence
print("After remove:", numbers)
popped = numbers.pop() # remove last and return it
print("After pop:", numbers, "Popped:", popped)
```

```python
fruits = ("apple", "banana", "cherry", "date", "elderberry")

# Indexing
print("First fruit:", fruits[0])
print("Last fruit:", fruits[-1])
print("Second fruit:", fruits[1])
```

```python
# Slicing
print("\nFirst three fruits:", fruits[0:3])
print("From index 2 onwards:", fruits[2:])
print("Up to index 3:", fruits[:3])
print("Every second fruit:", fruits[::2])
print("Reverse tuple:", fruits[::-1])
```

```python

# Dictionaries store key-value pairs for fast lookup
student = {'name': 'Alice', 'age': 25, 'grade': 'A'}

print("\nDictionary Example:", student)

# Accessing values
print("Student Name:", student['name'])
print("Student Age:", student['age'])

# Adding/updating values
student['age'] = 26
student['course'] = 'Data Science'
print("Updated Dictionary:", student)

# Dictionary methods
print("\nDictionary Methods:")
print("Keys:", student.keys())
print("Values:", student.values())
print("Items:", student.items())
```

```python
# Sets are unordered collections of unique elements; good for deduplication
my_set = {1, 2, 3, 3, 2}
print("\nSet Example:", my_set)

# Creating a set from a list
from_list = set([1, 2, 2, 3, 4])
print("Set from list (duplicates removed):", from_list)

# Set operations
A = {1, 2, 3}
B = {3, 4, 5}

print("Union:", A | B)       # combine unique elements
print("Intersection:", A & B) # common elements
print("Difference:", A - B)  # elements in A not in B
```
