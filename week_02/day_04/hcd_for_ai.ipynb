# 🧠 Human-Centered Design (HCD) for AI

> **Target outcome:** Stop building solutions for yourself and start designing for humans.

---

## The Problem HCD Solves

Here's a pattern that plays out constantly in AI projects:

> *A data science team spends three months building a churn prediction model. It achieves 94% accuracy. It gets deployed. The sales team never uses it.*

Why? Because the team asked **"What can we build with this data?"** instead of **"What does the sales team actually need?"**

The model was technically impressive but practically useless — it output a probability score with no explanation, delivered in a format the sales team couldn't read, about customers they couldn't action in time.

**Human-Centered Design** is the discipline that prevents this failure. It's a framework that ensures AI solutions are grounded in human needs rather than just technical possibilities.

## 1. The HCD Loop

HCD is not a linear process — it's a **loop** you repeat until you get it right.

```
                    ┌──────────────┐
              ┌────►│  EMPATHIZE   │◄────┐
              │     │  (Research)  │     │
              │     └──────┬───────┘     │
              │            │             │
         iterate           ▼         iterate
              │     ┌──────────────┐     │
              │     │    DEFINE    │     │
              │     │  (Synthesis) │     │
              │     └──────┬───────┘     │
              │            │             │
              │            ▼             │
              │     ┌──────────────┐     │
         ◄────┤     │   IDEATE     │     ├────►
              │     │(Brainstorm)  │     │
              │     └──────┬───────┘     │
              │            │             │
              │            ▼             │
              │     ┌──────────────┐     │
              └─────│  PROTOTYPE   │     │
                    │  (Build it)  │─────┘
                    └──────┬───────┘
                           │
                           ▼
                    ┌──────────────┐
                    │    TEST      │
                    │ (Get feedback│
                    │  & iterate)  │
                    └──────────────┘
```

### Each Stage Explained

**🔍 Empathize** — Understand the people you're designing for. Not through surveys with checkboxes, but through conversations, observations, and genuine curiosity. The goal is to feel what they feel.

**🎯 Define** — Synthesise your research into a clear problem statement. Not "users want a dashboard" but "community health workers need to know which patients to call today, ranked by urgency, without needing to log into 3 different systems."

**💡 Ideate** — Brainstorm solutions widely before narrowing. The rule: *quantity before quality.* Generate 20 ideas before judging any of them. Some of the best solutions are not AI at all.

**🔧 Prototype** — Build the simplest possible version that can be tested. For AI, this might be a paper mockup, a spreadsheet, or a simple rule-based system — before writing a single ML model.

**🧪 Test** — Put the prototype in front of real users and observe. Don't explain — watch. What confuses them tells you more than what they say.

## 2. Needs vs. Capabilities — The Innovation Sweet Spot

Every successful AI product lives at the intersection of three questions:

```
         DESIRABILITY
         (Do users want this?)
               ▲
               │
               │      ★
               │   Innovation
               │   Sweet Spot
               │
FEASIBILITY ───┼─────────────── VIABILITY
(Can we build  │               (Does it
 it with our   │                make sense
 data & tech?) │                for the org?)
```

| Question | What failure looks like |
|----------|-----------------------|
| **Desirability** only | Beautiful product nobody needed |
| **Feasibility** only | Technical demo nobody uses |
| **Viability** only | Profitable but harmful or unusable |

### The Churn Example Unpacked

| Dimension | Technical Team's View | User's View |
|-----------|----------------------|-------------|
| **Desirability** | "We predict churn at 94% accuracy" | "I need to know which customer to call TODAY and what to say" |
| **Feasibility** | "We have 3 years of transaction data" | "I check my phone between calls — it must be mobile-friendly" |
| **Viability** | "Model runs in our data warehouse" | "My manager needs to approve retention offers before I make the call" |

The sweet spot is a **mobile notification** that says: *"Customer Fatima — high churn risk. She hasn't used the app in 14 days. Offer code: RETAIN20 (approved)."*

Not a probability score. Not a dashboard. An **actionable nudge.**

## 3. User Research Fundamentals

Research is the foundation of HCD. Without it, you're guessing.

### Method 1: Interviews

**Goal:** Understand motivations, behaviours, and pain points — not preferences.

**The key rule:** Ask *why*, not *what*.

| ❌ What to Avoid | ✅ What to Ask Instead |
|-----------------|----------------------|
| "Would you use a dashboard?" | "Tell me about the last time you had to make a decision about a patient." |
| "Do you want AI recommendations?" | "Walk me through what happened the last time something went wrong." |
| "Is the current system bad?" | "What's the most frustrating part of your week? Why?" |

**The 5 Whys technique:**
```
User: "The reporting is frustrating."
You: "Why?"
User: "It takes too long."
You: "Why does it take long?"
User: "I have to pull data from three systems."
You: "Why three systems?"
User: "They were never integrated."
You: "Why not?"
User: "Nobody ever asked for it."
                      ↑
               REAL PROBLEM FOUND
```

### Method 2: Contextual Inquiry (Observation)

Watch users do their actual work in their actual environment. People do things differently from how they *describe* doing things.

**What to look for:**
- **Workarounds** — spreadsheets, sticky notes, WhatsApp groups = unmet needs
- **Moments of hesitation** — where do they slow down or re-check?
- **Context** — are they distracted? rushing? on a low-end phone?

### Finding Extreme Users

Extreme users — people who use your tool or face your problem in an unusual way — reveal edge cases and hidden needs that mainstream users take for granted.

| Mainstream User | Extreme User | Insight Gained |
|----------------|-------------|---------------|
| Health worker in a city clinic | Health worker with no internet in a rural area | Offline-first becomes essential |
| Literate adult | User with low literacy | Visual communication matters more than text |
| Manager who reviews AI decisions | User who must act on them in 2 minutes | Explanation needs to be immediate |


````python
# ============================================================
# INTERVIEW ANALYSIS TOOL
# Paste raw interview notes below — the tool helps you
# identify themes, pain points, and needs.
# ============================================================

from collections import Counter
import re

# Sample interview notes (replace with your own)
interview_notes = [
    {
        'participant': 'Amara, Community Health Worker, 5 years experience',
        'notes': '''
        - Visits 15 patients per day, mostly on foot
        - Uses a paper register to track visits — no smartphone data entry
        - Biggest frustration: doesn't know which patients need urgent follow-up until she checks the paper
        - Said: "Sometimes I walk to someone's house and they were fine, but someone else who needed me stayed home"
        - Workaround: calls supervisor by phone to decide who to visit next
        - Phone has 2G internet only
        - Pain point: reporting takes 2 hours every Friday
        '''
    },
    {
        'participant': 'Dr. Kofi, District Health Officer',
        'notes': '''
        - Manages 40 health workers across the district
        - Receives weekly paper reports — data is always outdated by 7 days when he sees it
        - Pain point: cannot identify which zones are underperforming until problem is serious
        - Wants: real-time view of coverage across his district
        - Concern: if AI gives wrong recommendation, he worries about accountability
        - Has laptop, good internet at office, but visits sites on a tablet
        '''
    },
    {
        'participant': 'Fatima, Patient, mother of 3',
        'notes': '''
        - Doesn't always remember appointment dates
        - Would prefer SMS reminders in Hausa, not English
        - Trusted her health worker more than the clinic — felt the clinic was far and expensive
        - Concern: doesn't want her health information shared without permission
        - Has a basic phone, no data plan, can receive SMS
        '''
    }
]

# Tag pain points automatically
pain_keywords = ['frustration', 'frustrated', 'pain', 'problem', 'wrong', 'worry',
                 'concern', 'difficult', 'hard', 'bad', "doesn't", 'cannot', "can't"]

need_keywords = ['want', 'need', 'would like', 'prefer', 'wish', 'should', 'better if']

print('=' * 60)
print('  INTERVIEW ANALYSIS SUMMARY')
print('=' * 60)

all_pain_points = []
all_needs = []

for interview in interview_notes:
    print(f'\n👤 {interview["participant"]}')
    print('-' * 50)
    lines = [l.strip('- ').strip() for l in interview['notes'].strip().split('\n') if l.strip('- ').strip()]

    pain_points = [l for l in lines if any(kw in l.lower() for kw in pain_keywords)]
    needs = [l for l in lines if any(kw in l.lower() for kw in need_keywords)]
    workarounds = [l for l in lines if 'workaround' in l.lower() or 'calls' in l.lower()]

    if pain_points:
        print('  🔴 Pain Points:')
        for p in pain_points:
            print(f'     → {p}')
            all_pain_points.append(p)
    if needs:
        print('  💚 Needs/Wants:')
        for n in needs:
            print(f'     → {n}')
            all_needs.append(n)

print(f'\n\nTotal pain points identified: {len(all_pain_points)}')
print(f'Total needs identified: {len(all_needs)}')
print('\n💡 Next step: Use these in the Define stage to write a Problem Statement.')
````

## 4. User Personas — Building Them Properly

A **persona** is a fictionalised but research-grounded representation of a real user type. It's NOT a demographic profile. It's a behavioural archetype.

### What Makes a Good Persona?

| ❌ Demographic Profile (useless) | ✅ Behavioural Persona (useful) |
|--------------------------------|--------------------------------|
| "Female, 28, urban, university education" | Goals: knows exactly who needs help, not waste time on wrong visits |
| "Income: ₦45,000/month" | Frustrations: data is always out of date when decisions are needed |
| "Uses smartphone" | Behaviours: texts colleagues informally to coordinate; checks paper register daily |
| — | Constraints: 2G internet, shares phone with family, limited battery |

### The Persona Template

Each persona should include:
1. **Name + Role** — humanising, memorable
2. **Goals** — what are they trying to achieve?
3. **Frustrations** — what stops them?
4. **Behaviours** — how do they actually work day-to-day?
5. **Constraints** — technology, literacy, time, connectivity
6. **Quote** — one sentence in their voice that captures their perspective

## 5. Journey Mapping — Finding Where AI Helps Most

A **journey map** visualises the full experience a user goes through to accomplish a goal — from start to finish. It reveals:
- Where users get frustrated (**"Moan Moments"** 😤)
- Where they feel confident (**"Yay Moments"** 😊)
- Where AI could provide the most relief

### Journey Map Structure

```
User: Amara (CHW)   Goal: Visit the right patients today

STAGE:    Wake Up    →  Check Register  →  Plan Route  →  Travel  →  Visit  →  Report
ACTIONS:  Gets ready    Reads paper        Guesses from  Walks to   Patient    Writes on
                        register           memory        homes      care       paper
FEELING:  😊 Energised  😐 Uncertain       😤 Frustrated  😓 Tired    😊 Helpful  😤 Tedious
                                             ↑                                    ↑
                                        MOAN MOMENT                          MOAN MOMENT
                                        [AI opportunity]                     [AI opportunity]
```

The **Moan Moments** are where your AI should target its energy.

````python
# ============================================================
# JOURNEY MAP VISUALISER
# ============================================================
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

# Define the journey — edit for YOUR persona and use case
journey = {
    'persona': 'Amara — Community Health Worker',
    'goal': 'Visit the right patients today and submit report',
    'stages': [
        'Start Day',
        'Plan Route',
        'Travel',
        'Patient Visit',
        'Re-Plan',
        'End of Day Report',
    ],
    # Emotion score: 5=very happy, 3=neutral, 1=very frustrated
    'emotion_score': [4, 1.5, 2, 5, 2, 1.5],
    'actions': [
        'Gets ready, checks paper register',
        'Guesses visit priority from memory',
        'Walks on foot — no info updates',
        'Provides care and records on paper',
        'Calls supervisor to re-prioritise',
        'Manually fills paper report (2 hrs)',
    ],
    'thoughts': [
        'Who needs me most today?',
        'I hope this is the right choice',
        'What if someone else needed me?',
        'I love this part of my job',
        'This call takes time I don\'t have',
        'This paperwork is the same data twice',
    ],
    'ai_opportunity': [
        False,
        True,  # ← AI: prioritised visit list
        True,  # ← AI: mid-day alert if urgent patient needs help
        False,
        True,  # ← AI: auto-suggest next patient
        True,  # ← AI: auto-populate report from visit data
    ]
}

fig, (ax_emo, ax_info) = plt.subplots(2, 1, figsize=(14, 8), gridspec_kw={'height_ratios': [2, 1]})
fig.suptitle(f'Journey Map: {journey["persona"]}\nGoal: {journey["goal"]}',
             fontsize=13, fontweight='bold', y=1.01)

stages = journey['stages']
x = np.arange(len(stages))
scores = journey['emotion_score']

# Emotion curve
ax_emo.plot(x, scores, 'o-', color='#028090', lw=3, markersize=10, zorder=5)
ax_emo.fill_between(x, scores, 3, where=[s < 3 for s in scores],
                     color='#E76F51', alpha=0.25, label='Pain zone')
ax_emo.fill_between(x, scores, 3, where=[s >= 3 for s in scores],
                     color='#028090', alpha=0.15, label='Positive zone')
ax_emo.axhline(3, color='gray', lw=1, linestyle='--', alpha=0.5, label='Neutral')

# Mark AI opportunities
for i, has_opp in enumerate(journey['ai_opportunity']):
    if has_opp:
        ax_emo.annotate('🤖 AI\nopportunity', xy=(i, scores[i]),
                        xytext=(i, scores[i] + 0.6 if scores[i] < 3 else scores[i] - 0.9),
                        ha='center', fontsize=8, color='#264653', fontweight='bold',
                        arrowprops=dict(arrowstyle='->', color='#264653', lw=1.5))

ax_emo.set_xticks(x)
ax_emo.set_xticklabels(stages, fontsize=11, fontweight='bold')
ax_emo.set_yticks([1, 2, 3, 4, 5])
ax_emo.set_yticklabels(['😤 Very\nFrustrated', '😓 Frustrated', '😐 Neutral', '😊 Positive', '🌟 Delighted'])
ax_emo.set_ylim(0.5, 6)
ax_emo.set_title('Emotional Journey', fontweight='bold')
ax_emo.legend(loc='upper right', fontsize=9)
ax_emo.spines['top'].set_visible(False)
ax_emo.spines['right'].set_visible(False)

# Actions and thoughts table
ax_info.axis('off')
col_width = 1 / len(stages)
for i, (action, thought) in enumerate(zip(journey['actions'], journey['thoughts'])):
    x_pos = i / len(stages) + col_width / 2
    color = '#E76F5133' if journey['ai_opportunity'][i] else 'white'
    rect = mpatches.FancyBboxPatch((i / len(stages) + 0.01, 0.05), col_width - 0.02, 0.9,
                                    boxstyle='round,pad=0.02', linewidth=1,
                                    edgecolor='#cccccc', facecolor=color,
                                    transform=ax_info.transAxes)
    ax_info.add_patch(rect)
    ax_info.text(x_pos, 0.72, action, ha='center', va='center', fontsize=7.5,
                 wrap=True, transform=ax_info.transAxes, style='italic',
                 color='#264653')
    ax_info.text(x_pos, 0.3, f'💭 "{thought}"', ha='center', va='center', fontsize=7,
                 wrap=True, transform=ax_info.transAxes, color='#555')

ax_info.text(-0.01, 0.72, 'Action:', ha='right', va='center',
             transform=ax_info.transAxes, fontweight='bold', fontsize=9)
ax_info.text(-0.01, 0.3, 'Thought:', ha='right', va='center',
             transform=ax_info.transAxes, fontweight='bold', fontsize=9)

orange_patch = mpatches.Patch(color='#E76F5133', label='AI opportunity stage')
ax_info.legend(handles=[orange_patch], loc='lower right', fontsize=8)

plt.tight_layout()
plt.show()

print()
print('🎯 AI Opportunity Summary:')
for i, (stage, has_opp, action) in enumerate(zip(stages, journey['ai_opportunity'], journey['actions'])):
    if has_opp:
        print(f'  Stage "{stage}": {action}')
````

## 6. Ethical Design & Responsibility

Every AI system makes choices — about what data to use, what to optimise for, who to surface, who to ignore. Those choices have consequences. Ethical design makes those choices deliberately and accountably.

### The Four Pillars

**♿ Accessibility**
Can every intended user actually use this? Consider:
- Screen readers and visual impairments
- Low-literacy users (rely on icons and audio, not text)
- Slow or unreliable internet (can it work offline?)
- Old or low-spec devices

**🌍 Inclusivity**
Does the training data fairly represent the people being affected? A model trained on urban patient data may perform poorly on rural patients. A model trained on predominantly one ethnic group may misclassify others.

**🤝 Cultural Sensitivity**
Does the AI's language, tone, and recommendations make sense across different cultural contexts? A recommendation that works in Lagos may be offensive or inappropriate in a different region.

**🔍 Transparency (Explainable AI)**
Can the user understand *why* the AI made a specific recommendation? Without explainability:
- Users can't catch errors
- Trust breaks down when the AI is wrong
- Accountability becomes impossible

> *"A 94% accurate black box is less useful than an 85% accurate model that explains its reasoning."*

### The Harm Checklist

Before deploying any AI feature, ask:

| Question | Risk if Ignored |
|----------|-----------------|
| Who is NOT in our training data? | Model fails for excluded groups |
| What happens when the model is wrong? | Who bears the cost of the error? |
| Can users override the AI's recommendation? | Loss of human agency |
| Is consent obtained for using this data? | Privacy violation |
| Does the model encode historical bias? | Discrimination at scale |
| Can the affected person see and contest the AI's decision? | Unaccountable systems |
| Who is responsible when something goes wrong? | Diffused accountability |

````python
# ============================================================
# ETHICAL DESIGN SCORECARD — Rate your AI project
# Score each dimension: 1 (not addressed) to 5 (fully addressed)
# ============================================================
import matplotlib.pyplot as plt
import numpy as np

# Fill in scores for YOUR project (1-5 for each)
ethics_scores = {
    'Accessibility\n(all users can use it)': 3,
    'Inclusivity\n(training data coverage)': 2,
    'Transparency\n(explainability)': 4,
    'Cultural\nSensitivity': 3,
    'Consent &\nPrivacy': 4,
    'Human\nOverride': 5,
    'Accountability\n(who\'s responsible?)': 3,
    'Bias\nMitigation': 2,
}

categories = list(ethics_scores.keys())
values = list(ethics_scores.values())
n = len(categories)

# Radar chart
angles = np.linspace(0, 2 * np.pi, n, endpoint=False).tolist()
values_plot = values + values[:1]
angles_plot = angles + angles[:1]

fig, (ax_radar, ax_bar) = plt.subplots(1, 2, figsize=(14, 6))

# Radar
ax_radar = plt.subplot(121, polar=True)
ax_radar.plot(angles_plot, values_plot, 'o-', color='#028090', lw=2)
ax_radar.fill(angles_plot, values_plot, color='#028090', alpha=0.25)
ax_radar.set_xticks(angles)
ax_radar.set_xticklabels(categories, size=9)
ax_radar.set_ylim(0, 5)
ax_radar.set_yticks([1, 2, 3, 4, 5])
ax_radar.set_yticklabels(['1', '2', '3', '4', '5'], size=8)
ax_radar.set_title('Ethical Design Scorecard', fontweight='bold', pad=20)

# Bar chart with action priority
ax_bar = plt.subplot(122)
colors = ['#E76F51' if v <= 2 else '#F4A261' if v == 3 else '#028090' for v in values]
bars = ax_bar.barh(range(n), values, color=colors, alpha=0.85, edgecolor='white')
ax_bar.set_yticks(range(n))
ax_bar.set_yticklabels([c.replace('\n', ' ') for c in categories], fontsize=9)
ax_bar.set_xlim(0, 5.5)
ax_bar.set_title('Priority Action Areas\n(Red = needs urgent attention)', fontweight='bold')
ax_bar.axvline(3, color='gray', lw=1, linestyle='--', alpha=0.5)

for bar, val in zip(bars, values):
    label = '⚠️ Action needed' if val <= 2 else '📋 Review' if val == 3 else '✅ Good'
    ax_bar.text(val + 0.1, bar.get_y() + bar.get_height() / 2,
                f'{val}/5  {label}', va='center', fontsize=8)

ax_bar.spines['top'].set_visible(False)
ax_bar.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()

print('\nAction Plan:')
urgent = [(c.replace('\n', ' '), v) for c, v in ethics_scores.items() if v <= 2]
if urgent:
    print('  🔴 URGENT — address before deployment:')
    for c, v in urgent:
        print(f'     • {c} (score: {v}/5)')
review = [(c.replace('\n', ' '), v) for c, v in ethics_scores.items() if v == 3]
if review:
    print('  🟡 REVIEW — plan improvements:')
    for c, v in review:
        print(f'     • {c} (score: {v}/5)')
````

## ✅ Summary

| HCD Stage | Key Question | Output |
|-----------|-------------|--------|
| **Empathize** | Who are we designing for? What do they actually experience? | Interview notes, observations |
| **Define** | What is the real problem? | Problem statement |
| **Ideate** | What could we build? (many options) | Idea list |
| **Prototype** | What's the simplest version to test? | Mockup, wireframe, or paper prototype |
| **Test** | Does it actually work for real users? | Feedback, iteration plan |

### The Most Important Mindset Shifts

1. **From "What can we build?" → "What should we build?"** — Technology follows human need, not the reverse.

2. **From "Users want X" → "Users are trying to do Y"** — People don't want a drill, they want a hole in the wall.

3. **From "Accuracy" → "Actionability"** — A 94% accurate model nobody acts on is less valuable than an 80% model with a clear call to action.

4. **From "Launch" → "Iterate"** — No design is right the first time. Build the loop into your project plan.
