### Solution 1: Generate Synthetic Sales Data

````python
import numpy as np

np.random.seed(42)

# Product parameters
products = ['Tomato', 'Onion', 'Pepper', 'Rice', 'Beans']
mean_sales = [100, 80, 60, 120, 90]
std_dev = 15

# Generate data: 90 days × 5 products
sales_data = np.zeros((90, 5))
for i, mean in enumerate(mean_sales):
    sales_data[:, i] = np.random.normal(mean, std_dev, size=90)
    # Ensure no negative sales
    sales_data[:, i] = np.maximum(sales_data[:, i], 0)

print(f"Sales data shape: {sales_data.shape}")
print(f"\nFirst 5 days:")
print(sales_data[:5])
print(f"\nData type: {sales_data.dtype}")
````

### Solution 2: Statistical Analysis

````python
# 1. Total sales per product
total_sales = sales_data.sum(axis=0)
print("1. Total sales per product (90 days):")
for product, total in zip(products, total_sales):
    print(f"   {product:8s}: {total:.1f} units")
print()

# 2. Average daily sales
avg_sales = sales_data.mean(axis=0)
print("2. Average daily sales:")
for product, avg in zip(products, avg_sales):
    print(f"   {product:8s}: {avg:.1f} units/day")
print()

# 3. Highest variability
std_sales = sales_data.std(axis=0)
max_var_idx = std_sales.argmax()
print("3. Product with highest sales variability:")
print(f"   {products[max_var_idx]}: σ = {std_sales[max_var_idx]:.2f} units")
print()

# 4. Day with maximum total sales
daily_totals = sales_data.sum(axis=1)
max_day = daily_totals.argmax()
print(f"4. Day with maximum total sales:")
print(f"   Day {max_day + 1}: {daily_totals[max_day]:.1f} units")
print()

# 5. Weekly average (reshape to 13 weeks × 7 days × 5 products, then average over days)
# Note: We'll use first 91 days to make it exactly 13 weeks
weekly_data = sales_data[:91].reshape(13, 7, 5)
weekly_avg = weekly_data.mean(axis=1)
print("5. Weekly average sales (first 13 weeks):")
print(f"   Shape: {weekly_avg.shape}")
print(f"\n   Week 1 averages:")
for product, avg in zip(products, weekly_avg[0]):
    print(f"   {product:8s}: {avg:.1f} units/day")
````

### Solution 3: Price Calculations

````python
# 1. Create price array
prices = np.array([250, 300, 400, 500, 350])
print("Prices:", prices)
print()

# 2. Calculate daily revenue
daily_revenue = sales_data @ prices  # Matrix multiplication
print("Daily revenue (first 10 days):")
print(daily_revenue[:10])
print(f"\nTotal revenue (90 days): ₦{daily_revenue.sum():,.2f}")
print()

# 3. Apply weekend discounts
# Create discount array: 0.8 for weekends, 1.0 for weekdays
discount = np.ones(90)
# Assuming Day 0 is Monday, weekends are days 5,6,12,13,19,20...
# Actually, let's use modulo: days 5,6 of each week
for i in range(90):
    if i % 7 in [5, 6]:  # Saturday, Sunday
        discount[i] = 0.8  # 20% discount

discounted_revenue = daily_revenue * discount
print("3. Revenue with weekend discounts:")
print(f"   Original total: ₦{daily_revenue.sum():,.2f}")
print(f"   Discounted total: ₦{discounted_revenue.sum():,.2f}")
print(f"   Total discount given: ₦{(daily_revenue.sum() - discounted_revenue.sum()):,.2f}")
print(f"   Number of weekend days: {(discount < 1).sum()}")
````

### Solution 4: Broadcasting Challenge

````python
sales = np.array([
    [10000, 8000, 12000],
    [15000, 12000, 14000],
    [9000, 7000, 11000]
])

# Tax rates for Lagos, Kano, Abuja
tax_rates = np.array([0.075, 0.05, 0.075])

# Broadcasting: add tax to each region (column-wise)
# Method: multiply sales by (1 + tax_rate)
sales_with_tax = sales * (1 + tax_rates)

print("Original sales:")
print(sales)
print()

print("Tax rates: [Lagos: 7.5%, Kano: 5.0%, Abuja: 7.5%]")
print()

print("Sales including tax:")
print(sales_with_tax)
print()

# Total sales per region
total_per_region = sales_with_tax.sum(axis=0)
regions = ['Lagos', 'Kano', 'Abuja']
print("Total sales (with tax) per region:")
for region, total in zip(regions, total_per_region):
    print(f"   {region:8s}: ₦{total:,.2f}")
````

### Solution 6: Load and Explore Market Data

````python
# 1. Load data
market_df = pd.read_csv('nigerian_market_prices.csv')

# 2. First and last 5 rows
print("First 5 rows:")
print(market_df.head())
print("\n" + "="*70 + "\n")

print("Last 5 rows:")
print(market_df.tail())
print("\n" + "="*70 + "\n")

# 3. Dataset information
print(f"Shape: {market_df.shape}")
print(f"Columns: {list(market_df.columns)}")
print(f"\nData types:")
print(market_df.dtypes)
print("\n" + "="*70 + "\n")

# 4. Statistical summary
print("Statistical summary:")
print(market_df.describe())
print("\n" + "="*70 + "\n")

# 5. Missing values
print("Missing values per column:")
print(market_df.isnull().sum())
print(f"\nTotal missing values: {market_df.isnull().sum().sum()}")
````

### Solution 7: Filtering and Selection

````python
# 1. Price between 200 and 300
filtered_price = market_df[(market_df['price'] >= 200) & (market_df['price'] <= 300)]
print("1. Records with price between ₦200-300:")
print(f"   Count: {len(filtered_price)}")
print(filtered_price.head())
print()

# 2. Lagos and Kano only
lagos_kano = market_df[market_df['state'].isin(['Lagos', 'Kano'])]
print("2. Records from Lagos and Kano:")
print(f"   Count: {len(lagos_kano)}")
print(lagos_kano.head())
print()

# 3. Top 10 most expensive
top_10_expensive = market_df.nlargest(10, 'price')
print("3. Top 10 most expensive products:")
print(top_10_expensive[['state', 'product', 'price']])
print()

# 4. Quantity > 50 AND price < 250
filtered_both = market_df[(market_df['quantity'] > 50) & (market_df['price'] < 250)]
print("4. Records with quantity > 50 AND price < ₦250:")
print(f"   Count: {len(filtered_both)}")
print()

# 5. Select specific columns
selected_cols = filtered_both[['state', 'product', 'price']]
print("5. Selected columns only:")
print(selected_cols.head())
````

### Solution 8: Aggregation and Grouping

````python
# 1. Average price per product
avg_by_product = market_df.groupby('product')['price'].mean().sort_values(ascending=False)
print("1. Average price per product:")
print(avg_by_product)
print()

# 2. Total quantity by state
qty_by_state = market_df.groupby('state')['quantity'].sum().sort_values(ascending=False)
print("2. Total quantity sold by state:")
print(qty_by_state)
print()

# 3. Min and max price by state
price_range = market_df.groupby('state')['price'].agg(['min', 'max'])
print("3. Price range by state:")
print(price_range)
print()

# 4. Median price for each product-state combination
median_combo = market_df.groupby(['product', 'state'])['price'].median()
print("4. Median price by product-state (first 10):")
print(median_combo.head(10))
print()

# 5. Count records per state
count_by_state = market_df['state'].value_counts()
print("5. Number of records per state:")
print(count_by_state)
````

### Solution 9: Data Transformation

````python
# Make a copy to avoid modifying original
market_transformed = market_df.copy()

# 1. Total value
market_transformed['total_value'] = market_transformed['price'] * market_transformed['quantity']
print("1. Added 'total_value' column:")
print(market_transformed[['price', 'quantity', 'total_value']].head())
print()

# 2. Price category
def categorize_price(price):
    if price < 200:
        return 'Low'
    elif price <= 350:
        return 'Medium'
    else:
        return 'High'

market_transformed['price_category'] = market_transformed['price'].apply(categorize_price)
# Alternative using pd.cut:
# market_transformed['price_category'] = pd.cut(
#     market_transformed['price'],
#     bins=[0, 200, 350, float('inf')],
#     labels=['Low', 'Medium', 'High']
# )

print("2. Added 'price_category' column:")
print(market_transformed[['price', 'price_category']].head(10))
print(f"\nCategory distribution:")
print(market_transformed['price_category'].value_counts())
print()

# 3. Convert to USD
market_transformed['price_usd'] = market_transformed['price'] / 1500
print("3. Added 'price_usd' column:")
print(market_transformed[['price', 'price_usd']].head())
print()

# 4. Above average indicator
mean_price = market_transformed['price'].mean()
market_transformed['above_average'] = market_transformed['price'] > mean_price
print(f"4. Added 'above_average' column (mean price: ₦{mean_price:.2f}):")
print(market_transformed[['price', 'above_average']].head(10))
print(f"\nRecords above average: {market_transformed['above_average'].sum()}")
````

### Solution 10: Population Analysis

````python
# 1. Load population data
pop_df = pd.read_csv('nigerian_population_data.csv')
print("Population data loaded:")
print(pop_df.head())
print()

# 2. Population density
pop_df['density'] = pop_df['population'] / pop_df['area_km2']
print("2. Population density added:")
print(pop_df[['state', 'population', 'area_km2', 'density']].head())
print()

# 3. Sort by density
sorted_by_density = pop_df.sort_values('density', ascending=False)
print("3. States sorted by population density:")
print(sorted_by_density[['state', 'density']].head(10))
print()

# 4. Large area states
large_area = pop_df[pop_df['area_km2'] > 30000]
print("4. States with area > 30,000 km²:")
print(large_area[['state', 'area_km2']])
print()

# 5. Percentage of total population
total_population = pop_df['population'].sum()
pop_df['pct_of_total'] = (pop_df['population'] / total_population) * 100
print("5. Population percentage:")
print(pop_df[['state', 'population', 'pct_of_total']].sort_values('pct_of_total', ascending=False).head(10))
print()

# 6. Population categories
def categorize_population(pop):
    if pop < 3_000_000:
        return 'Small'
    elif pop <= 7_000_000:
        return 'Medium'
    else:
        return 'Large'

pop_df['size_category'] = pop_df['population'].apply(categorize_population)
print("6. Population size categories:")
print(pop_df['size_category'].value_counts())
print()
print("Examples:")
print(pop_df[['state', 'population', 'size_category']].head(10))
````

### Solution 11: Combining Datasets (Challenge!)

````python
# Aggregate market data by state
market_agg = market_df.groupby('state').agg({
    'price': 'mean',
    'quantity': 'sum'
}).reset_index()
market_agg.columns = ['state', 'avg_price', 'total_quantity']

# Add total value
market_agg['total_market_value'] = market_agg['avg_price'] * market_agg['total_quantity']

print("Aggregated market data:")
print(market_agg)
print()

# Merge with population data
merged = pd.merge(
    market_agg,
    pop_df[['state', 'population']],
    on='state',
    how='inner'
)

print("Merged dataset:")
print(merged)
print()

# 1. Correlation between population and prices
correlation = merged['population'].corr(merged['avg_price'])
print(f"1. Correlation between population and average price: {correlation:.3f}")
if correlation > 0.5:
    print("   → Strong positive correlation: high-population states tend to have higher prices")
elif correlation > 0:
    print("   → Weak positive correlation: slight tendency for higher prices in populous states")
elif correlation > -0.5:
    print("   → Weak negative correlation: slight tendency for lower prices in populous states")
else:
    print("   → Strong negative correlation: high-population states tend to have lower prices")
print()

# 2. Market value per capita
merged['value_per_capita'] = merged['total_market_value'] / merged['population']
print("2. Total market value per capita:")
print(merged[['state', 'value_per_capita']].sort_values('value_per_capita', ascending=False))
print()

# 3. Best value states (lowest prices)
best_value = merged.sort_values('avg_price')
print("3. States with best value (lowest average prices):")
print(best_value[['state', 'avg_price', 'population']].head())
print()

# Visualization of relationship
print("Price vs Population scatter:")
for _, row in merged.iterrows():
    print(f"{row['state']:12s}: Pop={row['population']:>10,} | Avg Price=₦{row['avg_price']:>6.2f}")
````
