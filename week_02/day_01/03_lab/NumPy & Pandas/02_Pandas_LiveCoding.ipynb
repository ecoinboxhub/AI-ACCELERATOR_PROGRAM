## 1. Introduction to Pandas

Pandas = **Pan**el **Da**ta + Python Data Analysis

Think of it as:
- Excel in Python (but much more powerful)
- Built on top of NumPy

````python
import pandas as pd

import numpy as np

# Check version
print(f"Pandas version: {pd.__version__}")
````

## 2. Series: One-Dimensional Data

A Series is like a single column in a spreadsheet.

````python
# Create a Series from a list
prices = pd.Series([250, 180, 220, 200, 190])
print("Simple Series:")
print(prices)
print()

# Series with custom index (labels)
prices_labeled = pd.Series(
    [250, 180, 220, 200, 190],
    index=['Lagos', 'Kano', 'Abuja', 'Rivers', 'Oyo'],
    name='Tomato_Price'
)

print("Labeled Series:")
print(prices_labeled)
print()

# Accessing elements
print("Price in Lagos:", prices_labeled['Lagos'])
print("Price in Kano:", prices_labeled['Kano'])
print()

# Operations on Series
print("Prices with 10% increase:")
print(prices_labeled * 1.1)
````

## 3. DataFrames: Two-Dimensional Data

A DataFrame is like a spreadsheet or SQL table.

````python
# Create DataFrame from dictionary
data = {
    'state': ['Lagos', 'Kano', 'Abuja', 'Rivers', 'Oyo'],
    'tomato_price': [250, 180, 220, 200, 190],
    'onion_price': [300, 250, 280, 270, 260],
    'population_million': [15.4, 13.8, 3.8, 7.3, 7.8]
}

df = pd.DataFrame(data)
print("Market Data DataFrame:")
print(df)
print()

# DataFrame info
print(f"Shape: {df.shape} (rows, columns)")
print(f"Columns: {list(df.columns)}")
print(f"Data types:\n{df.dtypes}")
````

## 4. Loading Data from CSV

Real-world data comes from files. Let's load our Nigerian datasets!

````python
# Load market prices data
market_df = pd.read_csv('nigerian_market_prices.csv')

print("Market Prices Dataset:")
print(market_df.head())  # First 5 rows
print()
print(f"Dataset shape: {market_df.shape}")
````

## 5. Essential Data Exploration Methods

Always explore your data before manipulating it!

````python
# Display first few rows
print("First 5 rows:")
print(market_df.head())
print("\n" + "="*60 + "\n")

# Display last few rows
print("Last 3 rows:")
print(market_df.tail(3))
print("\n" + "="*60 + "\n")

# Dataset info: types, non-null counts
print("Dataset Info:")
market_df.info()
print("\n" + "="*60 + "\n")

# Statistical summary
print("Statistical Summary:")
print(market_df.desc    ribe())
print("\n" + "="*60 + "\n")

# Column names
print("Columns:", list(market_df.columns))
print()

# Shape
print(f"Shape: {market_df.shape[0]} rows × {market_df.shape[1]} columns")
````

## 6. Column Selection

Selecting columns is one of the most common operations.

````python
# Single column (returns Series)
state_column = market_df['state']
print("State column (Series):")
print(state_column)
print(f"Type: {type(state_column)}")
print()

# Multiple columns (returns DataFrame)
subset = market_df[['state', 'product', 'price']]
print("Selected columns (DataFrame):")
print(subset.head())
print(f"Type: {type(subset)}")
print()

# Adding a new column
market_df['price_usd'] = market_df['price'] / 1500  # Convert to USD
print("With new column:")
print(market_df.head())
````

## 7. Row Selection: .loc vs .iloc

**Key Difference:**
- `.loc[]` = label-based (use actual index labels)
- `.iloc[]` = position-based (use integer positions)

````python
# .iloc - position-based indexing
print("First row using .iloc[0]:")
print(market_df.iloc[0])
print()

print("First 3 rows using .iloc[0:3]:")
print(market_df.iloc[0:3])
print()

# .loc - label-based indexing (also works with integer labels)
print("First row using .loc[0]:")
print(market_df.loc[0])
print()

print("Rows 0 to 2 using .loc[0:2]:")
print(market_df.loc[0:2])  # Note: inclusive on both ends!
print()

# Selecting specific rows and columns
print("Specific rows and columns:")
print(market_df.loc[0:2, ['state', 'price']])
````

## 8. Filtering Data (Boolean Indexing)

This is where Pandas becomes powerful!

````python
# Filter: prices above 200
high_prices = market_df[market_df['price'] > 200]
print("Products with price > ₦200:")
print(high_prices)
print()

# Filter: specific state
lagos_data = market_df[market_df['state'] == 'Lagos']
print("Lagos market data:")
print(lagos_data)
print()

# Multiple conditions (use & for AND, | for OR)
# Must use parentheses around each condition!
filtered = market_df[(market_df['price'] > 200) & (market_df['state'] == 'Lagos')]
print("Lagos products with price > ₦200:")
print(filtered)
print()

# OR condition
filtered_or = market_df[(market_df['state'] == 'Lagos') | (market_df['state'] == 'Kano')]
print("Data from Lagos OR Kano:")
print(filtered_or)
````

## 9. Sorting Data

````python
# Sort by price (ascending)
sorted_asc = market_df.sort_values('price')
print("Sorted by price (ascending):")
print(sorted_asc.head())
print()

# Sort by price (descending)
sorted_desc = market_df.sort_values('price', ascending=False)
print("Sorted by price (descending):")
print(sorted_desc.head())
print()

# Sort by multiple columns
sorted_multi = market_df.sort_values(['state', 'price'], ascending=[True, False])
print("Sorted by state (A-Z), then price (high to low):")
print(sorted_multi.head(10))
````

## 10. Basic Aggregations

Calculate summary statistics on your data.

````python
# Overall statistics
print("Average price:", market_df['price'].mean())
print("Median price:", market_df['price'].median())
print("Min price:", market_df['price'].min())
print("Max price:", market_df['price'].max())
print("Total quantity:", market_df['quantity'].sum())
print()

# Value counts: how many records per state?
print("Records per state:")
print(market_df['state'].value_counts())
print()

# Unique values
print("Unique states:", market_df['state'].unique())
print("Number of unique states:", market_df['state'].nunique())
````

## 11. GroupBy: Split-Apply-Combine

One of the most powerful Pandas operations!

````python
# Average price by state
avg_by_state = market_df.groupby('state')['price'].mean()
print("Average price by state:")
print(avg_by_state)
print()

# Multiple aggregations
agg_by_state = market_df.groupby('state').agg({
    'price': ['mean', 'min', 'max'],
    'quantity': 'sum'
})
print("Aggregated statistics by state:")
print(agg_by_state)
print()

# Group by multiple columns
by_state_product = market_df.groupby(['state', 'product'])['price'].mean()
print("Average price by state and product:")
print(by_state_product)
````

## 12. Practical Example: Nigerian Market Analysis

Let's analyze the market data to answer business questions.

````python
# Question 1: Which state has the highest average prices?
avg_prices = market_df.groupby('state')['price'].mean().sort_values(ascending=False)
print("States ranked by average price:")
print(avg_prices)
print(f"\nMost expensive state: {avg_prices.index[0]} (₦{avg_prices.iloc[0]:.2f})")
print()

# Question 2: What's the most expensive product?
product_prices = market_df.groupby('product')['price'].mean().sort_values(ascending=False)
print("Products ranked by average price:")
print(product_prices)
print()

# Question 3: Total market value by state (price × quantity)
market_df['total_value'] = market_df['price'] * market_df['quantity']
value_by_state = market_df.groupby('state')['total_value'].sum().sort_values(ascending=False)
print("Total market value by state:")
for state, value in value_by_state.items():
    print(f"{state:15s}: ₦{value:,.2f}")
print()

# Question 4: Find outliers (prices > 2 standard deviations from mean)
mean_price = market_df['price'].mean()
std_price = market_df['price'].std()
outliers = market_df[(market_df['price'] > mean_price + 2*std_price) | 
                      (market_df['price'] < mean_price - 2*std_price)]
print(f"Price outliers (mean ± 2σ: ₦{mean_price:.2f} ± {2*std_price:.2f}):")
print(outliers[['state', 'product', 'price']])
````

## 15. Combining Insights

Let's correlate market prices with population data.

````python
# Average market price by state
avg_market_price = market_df.groupby('state')['price'].mean().reset_index()
avg_market_price.columns = ['state', 'avg_price']

# Merge with population data
merged = pd.merge(avg_market_price, pop_df[['state', 'population']], on='state', how='inner')
print("Average market price vs Population:")
print(merged)
print()

# Correlation between population and prices
correlation = merged['population'].corr(merged['avg_price'])
print(f"Correlation between population and average price: {correlation:.3f}")

if abs(correlation) < 0.3:
    print("Weak correlation - population doesn't strongly affect prices")
elif abs(correlation) < 0.7:
    print("Moderate correlation - some relationship exists")
else:
    print("Strong correlation - population significantly affects prices")
````
