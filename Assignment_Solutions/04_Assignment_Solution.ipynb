## Part A: NumPy Array Operations (30 points)

### Question 1: Revenue Simulation (10 points)

You run a small Nigerian market stall selling 4 products. Simulate 30 days of business:

**Requirements:**
1. Generate daily sales quantities for 4 products (use normal distribution)
   - Product A: mean=50, std=10
   - Product B: mean=30, std=5
   - Product C: mean=70, std=15
   - Product D: mean=40, std=8
2. Set random seed to your favorite number
3. Ensure no negative quantities (clip at 0)
4. Print the shape and first 5 days

**Deliverable:** Create a 30×4 NumPy array called `daily_sales`

```python
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Generate sales data with specified distributions
product_a = np.random.normal(50, 10, 30)
product_b = np.random.normal(30, 5, 30)
product_c = np.random.normal(70, 15, 30)
product_d = np.random.normal(40, 8, 30)

# Stack arrays to create 30x4 array
daily_sales = np.column_stack([product_a, product_b, product_c, product_d])

# Ensure no negative quantities (clip at 0)
daily_sales = np.clip(daily_sales, 0, None)

# Display results
print("Shape of daily_sales:", daily_sales.shape)
print("\nFirst 5 days of sales:")
print(daily_sales[:5])
```

### Question 2: Financial Analysis (10 points)

Using the `daily_sales` array from Q1:

**Calculate:**
1. Total sales for each product over 30 days
2. Average daily sales per product
3. The day with maximum total sales (across all products)
4. How many days had total sales above the 30-day average?
5. Standard deviation of sales for each product

**Deliverable:** Print all results with clear labels

```python
# Question 2: Financial Analysis

# 1. Total sales for each product over 30 days
total_sales = np.sum(daily_sales, axis=0)
print("Total sales for each product over 30 days:")
print(f"Product A: {total_sales[0]:.2f}")
print(f"Product B: {total_sales[1]:.2f}")
print(f"Product C: {total_sales[2]:.2f}")
print(f"Product D: {total_sales[3]:.2f}")

# 2. Average daily sales per product
average_daily_sales = np.mean(daily_sales, axis=0)
print("\nAverage daily sales per product:")
print(f"Product A: {average_daily_sales[0]:.2f}")
print(f"Product B: {average_daily_sales[1]:.2f}")
print(f"Product C: {average_daily_sales[2]:.2f}")
print(f"Product D: {average_daily_sales[3]:.2f}")

# 3. The day with maximum total sales (across all products)
daily_totals = np.sum(daily_sales, axis=1)
max_day = np.argmax(daily_totals)
print(f"\nDay with maximum total sales: Day {max_day + 1} with {daily_totals[max_day]:.2f} units")

# 4. How many days had total sales above the 30-day average?
overall_average = np.mean(daily_totals)
days_above_average = np.sum(daily_totals > overall_average)
print(f"\nDays with total sales above 30-day average: {days_above_average} days")

# 5. Standard deviation of sales for each product
std_dev = np.std(daily_sales, axis=0)
print("\nStandard deviation of sales for each product:")
print(f"Product A: {std_dev[0]:.2f}")
print(f"Product B: {std_dev[1]:.2f}")
print(f"Product C: {std_dev[2]:.2f}")
print(f"Product D: {std_dev[3]:.2f}")
```

### Question 3: Price & Revenue Calculation (10 points)

Your products have these prices: [200, 350, 150, 280] NGN

**Tasks:**
1. Calculate daily revenue (price × quantity for each product)
2. Calculate total revenue over 30 days
3. Apply a promotional discount:
   - Days 1-5: 15% off all products
   - Days 15-20: 10% off Product A and B only
   - Other days: no discount
4. Calculate total revenue with discounts
5. How much discount did you give away?

**Hint:** Use boolean indexing and broadcasting

```python
# Question 3: Price & Revenue Calculation
prices = np.array([200, 350, 150, 280])

# 1. Calculate daily revenue (price × quantity for each product)
daily_revenue = daily_sales * prices
print("Daily Revenue for first 5 days:")
print(daily_revenue[:5])

# 2. Calculate total revenue over 30 days
total_revenue_no_discount = np.sum(daily_revenue)
print(f"\nTotal revenue without discount: ₦{total_revenue_no_discount:.2f}")

# 3. Apply promotional discount
discount_rate = np.zeros(30)
# Days 1-5: 15% off all products
discount_rate[0:5] = 0.15
# Days 15-20: 10% off Product A and B only
discount_rate[14:20] = 0.10

# Create a discount array (30 x 4) where day 15-20 has 10% for products A,B only
discount_array = np.zeros_like(daily_revenue, dtype=float)
discount_array[0:5, :] = 0.15  # 15% off all products for days 1-5
discount_array[14:20, 0:2] = 0.10  # 10% off Product A and B for days 15-20

# 4. Calculate revenue with discounts
revenue_with_discount = daily_revenue * (1 - discount_array)
total_revenue_with_discount = np.sum(revenue_with_discount)
print(f"Total revenue with discounts: ₦{total_revenue_with_discount:.2f}")

# 5. How much discount was given away?
total_discount_given = total_revenue_no_discount - total_revenue_with_discount
print(f"\nTotal discount given away: ₦{total_discount_given:.2f}")
print(f"Discount percentage: {(total_discount_given/total_revenue_no_discount*100):.2f}%")
```

### Question 7: Data Transformation (15 points)

Using your dataset:

**Tasks:**
1. Create a new column based on a calculation (e.g., total = price × quantity)
2. Create a categorical column based on numeric ranges (e.g., 'Low', 'Medium', 'High')
3. Sort data by 2 or more columns
4. Calculate a rolling or cumulative statistic (if time-series data)

```python
# Question 7: Data Transformation

df_transformed = df.copy()

# 1. Create a new column based on calculation (revenue = price × quantity)
df_transformed['revenue'] = df_transformed['price'] * df_transformed['quantity']
print("1. New calculated column 'revenue' (first 5 rows):")
print(df_transformed[['price', 'quantity', 'revenue']].head())

# 2. Create a categorical column based on numeric ranges
df_transformed['price_category'] = pd.cut(df_transformed['price'], 
                                           bins=[0, 250, 375, 530], 
                                           labels=['Low', 'Medium', 'High'])
print("\n2. Categorical column 'price_category':")
print(df_transformed[['price', 'price_category']].head(10))
print("\nPrice category distribution:")
print(df_transformed['price_category'].value_counts())

# 3. Sort data by price (descending) then quantity (descending)
print("\n3. Sorted data by price and quantity (first 10 rows):")
sorted_df = df_transformed.sort_values(by=['price', 'quantity'], ascending=[False, False]).head(10)
print(sorted_df[['state', 'product', 'price', 'quantity']])

# 4. Calculate cumulative sum and rolling mean for quantity by state
df_transformed['cumulative_qty'] = df_transformed.groupby('state')['quantity'].cumsum()
df_transformed['rolling_mean_qty'] = df_transformed.groupby('state')['quantity'].transform(
    lambda x: x.rolling(window=3, min_periods=1).mean()
)
print("\n4. Cumulative and rolling statistics (first 10 rows):")
print(df_transformed[['state', 'quantity', 'cumulative_qty', 'rolling_mean_qty']].head(10))

# 5. Handle missing values (check if any exist)
print("\n5. Missing values handling:")
missing_before = df_transformed.isnull().sum().sum()
print(f"Missing values before: {missing_before}")
# Use ffill() and bfill() instead of method parameter for numeric columns
numeric_cols = df_transformed.select_dtypes(include=['number']).columns
df_transformed[numeric_cols] = df_transformed[numeric_cols].ffill().bfill().fillna(0)
missing_after = df_transformed.isnull().sum().sum()
print(f"Missing values after: {missing_after}")

print("\nTransformed dataset shape:", df_transformed.shape)
print(f"New columns added: {[col for col in df_transformed.columns if col not in df.columns]}")
```

#### Question 8a: What is the distribution of prices across the market?

What is the distribution of prices across the market and which price range is most common?

```python
# Question 8a: Price distribution analysis
# Reload the original dataset to ensure fresh data
df_q8 = pd.read_csv('nigerian_market_prices.csv')
price_col = df_q8['price']

# Calculate statistics
print("Price Distribution Analysis:")
print(f"Minimum price: ₦{price_col.min():.2f}")
print(f"Maximum price: ₦{price_col.max():.2f}")
print(f"Mean price: ₦{price_col.mean():.2f}")
print(f"Median price: ₦{price_col.median():.2f}")
print(f"Standard deviation: ₦{price_col.std():.2f}")

# Create price bins
df_analysis = df_q8.copy()
df_analysis['price_bin'] = pd.cut(price_col, bins=5)

# Count prices in each bin
print("\nPrice range distribution:")
price_dist = df_analysis['price_bin'].value_counts().sort_index()
print(price_dist)

# Find most common price range
most_common_idx = price_dist.argmax()
most_common_range = price_dist.index[most_common_idx]
print(f"\nMost common price range: {most_common_range}")
print(f"Number of products: {price_dist.iloc[most_common_idx]}")
```

**Interpretation:**

The price distribution analysis reveals the central tendency and spread of market prices. The mean and median values indicate where most products are priced, while the standard deviation shows the variability in pricing. By identifying the most common price range, we can understand consumer preferences and market positioning. This information is crucial for pricing strategies and identifying market opportunities in different price segments.

#### Question 8b: How do products compare in terms of price volatility?

Which product category or type has the highest price volatility, and what does this tell us about market conditions?

```python
# Question 8b: Price volatility analysis
# Reload the original dataset
df_q8b = pd.read_csv('nigerian_market_prices.csv')
price_col = df_q8b['price']

# Calculate coefficient of variation (volatility measure)
mean_price = price_col.mean()
std_price = price_col.std()
coef_variation = (std_price / mean_price) * 100

print("Price Volatility Analysis:")
print(f"Coefficient of Variation: {coef_variation:.2f}%")

# Analyze by state
volatility_by_state = df_q8b.groupby('state')['price'].apply(
    lambda x: (x.std() / x.mean() * 100) if x.mean() != 0 else 0
).sort_values(ascending=False)

print(f"\nPrice volatility by state:")
print(volatility_by_state)

most_volatile = volatility_by_state.idxmax()
print(f"\nMost volatile state: {most_volatile} (CV: {volatility_by_state.iloc[0]:.2f}%)")
least_volatile = volatility_by_state.idxmin()
print(f"Most stable state: {least_volatile} (CV: {volatility_by_state.iloc[-1]:.2f}%)")
```

**Interpretation:**

Price volatility indicates market stability and predictability. High volatility suggests price fluctuations due to supply-demand imbalances, seasonal factors, or external shocks, making it risky for both consumers and sellers. Products with low volatility are more stable and predictable, which is favorable for planning and budgeting. Understanding these patterns helps traders anticipate price changes and make informed purchasing and selling decisions.

#### Question 8c: What are the key statistical insights about market data?

What are the main statistical insights and outliers in the market data that could influence business decisions?

```python
# Question 8c: Statistical insights and outliers
# Reload the original dataset
df_q8c = pd.read_csv('nigerian_market_prices.csv')
price_col = df_q8c['price']

# Calculate quartiles and IQR
Q1 = price_col.quantile(0.25)
Q3 = price_col.quantile(0.75)
IQR = Q3 - Q1

print("Statistical Insights:")
print(f"Q1 (25th percentile): ₦{Q1:.2f}")
print(f"Q3 (75th percentile): ₦{Q3:.2f}")
print(f"Interquartile Range: ₦{IQR:.2f}")

# Identify outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = (price_col < lower_bound) | (price_col > upper_bound)
num_outliers = outliers.sum()

print(f"\nOutlier Analysis:")
print(f"Lower bound: ₦{lower_bound:.2f}")
print(f"Upper bound: ₦{upper_bound:.2f}")
print(f"Number of outliers: {num_outliers}")
print(f"Percentage of outliers: {(num_outliers/len(df_q8c)*100):.2f}%")

if num_outliers > 0:
    print(f"\nOutlier values:")
    print(df_q8c[outliers][['state', 'product', 'price', 'quantity']].head())
else:
    print("\nNo outliers detected based on IQR method")

# Calculate skewness manually
mean = price_col.mean()
std = price_col.std()
n = len(price_col)
skewness = ((price_col - mean) ** 3).sum() / (n * std ** 3)

print(f"\nSkewness: {skewness:.3f}")
if abs(skewness) < 0.5:
    print("Distribution is fairly symmetric")
elif skewness > 0:
    print("Distribution is right-skewed (positive skew)")
else:
    print("Distribution is left-skewed (negative skew)")
```

**Interpretation:**

Statistical insights like quartiles, IQR, and skewness help us understand the data distribution and identify exceptional cases. Outliers represent unusual prices that deviate significantly from the norm—these could be premium products, data errors, or special circumstances. Understanding the skewness tells us whether prices tend to cluster toward the lower or higher end. These insights guide pricing strategies, help detect data quality issues, and inform risk management decisions in the market.

```python
# BONUS CHALLENGE: Option C - Custom Aggregation Function

# Reload the original dataset
df_bonus = pd.read_csv('nigerian_market_prices.csv')

# Define a custom function for custom aggregation
def calculate_price_metrics(group):
    """
    Custom function to calculate comprehensive price metrics for a group
    Returns min, max, mean, median, and range
    """
    return pd.Series({
        'min_price': group.min(),
        'max_price': group.max(),
        'mean_price': group.mean(),
        'median_price': group.median(),
        'price_range': group.max() - group.min(),
        'count': len(group)
    })

# Apply custom function using groupby and apply
bonus_result = df_bonus.groupby('state')['price'].apply(calculate_price_metrics)
print("BONUS Challenge Results - Custom Aggregation by State:")
print(bonus_result)

print("\n" + "="*60)
print("EXPLANATION:")
print("="*60)
print("""
Custom aggregation functions are needed when built-in pandas functions cannot 
provide the specific insights required. In this case, we created a function that 
calculates multiple complementary metrics (min, max, mean, median, range, count) 
in a single operation. This is more efficient than calling separate aggregation 
functions and provides a unified view of the data characteristics for each group. 
The custom function returns a Series, making it compatible with pandas' apply() method.
""")
```
